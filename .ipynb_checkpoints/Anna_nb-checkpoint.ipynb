```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# load dataset
data = pd.read_csv(r"C:\Users\Shinemet\Ironhack\Projects\Project7\Original files\imdb_processed.csv")
```

```python
data.tail()
```

```python
# dropping redundant first column
data.drop('Unnamed: 0', axis=1, inplace=True)
```

# unstacking columns

```python
col_to_clean = ['genre', 'country', 'language', 'cast', 'director', 'composer', 'writer']

for col in col_to_clean:
    data[col] = data[col].str.strip('[]')
```

```python
# Splitting each column into a maximum of 2 columns
```

```python
genre = data.genre.str.split(',').apply(pd.Series)
genre.drop(columns=[2,3,4,5,6,7,8,9,10], inplace=True)
genre.rename(columns={0: 'genre1', 1: 'genre2'}, inplace=True)
genre
```

```python
country = data.country.str.split(',').apply(pd.Series)
country.drop(columns=[2,3,4,5,6,7,8,9,10], inplace=True)
country.rename(columns={0: 'country1', 1: 'country2'}, inplace=True)
country
```

```python
language = data.language.str.split(',').apply(pd.Series)
language.drop(columns=[2,3,4,5,6,7,8,9,10], inplace=True)
country.rename(columns={0: 'country1', 1: 'country2'}, inplace=True)
country
```
